```python
# Name : MUHAMMAD AIMAN ARIF BIN KADAR 
# Matric number : 204199
# LAB 2 

import random 

list_1 = []
list_2 = []
list_3 = []

#########################################

# update function 
def update_list(myList) :
  for x in range(1,11):
    myList.append(random.randrange(1,1001))
  return myList

# sum function 
def sum_list(myList):
  s = sum(myList)
  return s 

# average function 
def average_list(sumV , length):
  avg = sumV / length 
  return avg 

# find max 
def find_max(myList):
  mx = max(myList)
  return mx

# find min 
def find_min(myList):
  mn = min(myList)
  return mn

# find number less then average 
def less_avg(myList , avrg):
  num = []
  for i in range(len(myList)):
    if myList[i] < avrg:
      x = myList[i]
      num.append(x) 
  return num

# find number greater than average 
def greater_avg(myList , avrg):
  num = []
  for i in range(len(myList)):
    if myList[i] > avrg:
      x = myList[i]
      num.append(x)
  return num

##########################################                  

# upadate list 
update_list(list_1) 
print("List 1 : " , list_1)
update_list(list_2) 
print("List 2 : " , list_2)
update_list(list_3) 
print("List 3 : " , list_3)
print("\n")

# sum the list 
su1 = sum_list(list_1)
print("Sum all list 1 : ",su1)
su2 = sum_list(list_2)
print("Sum all list 2 : " , su2)
su3 = sum_list(list_3)
print("Sum all list 3 : " , su3)
print("\n")

# average list 
avr1 = average_list(su1,len(list_1))
print("Average list 1 : ",avr1)
avr2 = average_list(su2,len(list_2))
print("Average list 2 : ",avr2)
avr3 = average_list(su3,len(list_3))
print("Average list 3 : ",avr3)
print("\n")

# max number in the list 
max_num_1 = find_max(list_1)
print("Maximum number of the list 1 : ",max_num_1)
max_num_2 = find_max(list_2)
print("Maximum number of the list 2 : ",max_num_2)
max_num_3 = find_max(list_3)
print("Maximum number of the list 3 : ",max_num_3)
print("\n")

# min number in the list 
min_num_1 = find_min(list_1)
print("Minimum number of the list 1  :",min_num_1)
min_num_2 = find_min(list_2)
print("Minimum number of the list 2  :",min_num_2)
min_num_3 = find_min(list_3)
print("Minimum number of the list 3  :",min_num_3)
print("\n")

# number less than average 
min_avg_1 = less_avg(list_1,avr1)
print("Number less than average (List 1 ) : ",min_avg_1)
min_avg_2 = less_avg(list_2,avr2)
print("Number less than average (List 2 ) : ",min_avg_2)
min_avg_3 = less_avg(list_3,avr3)
print("Number less than average (List 3 ) : ",min_avg_3)
print("\n")

# number greater than average 
mx_avg_1 = greater_avg(list_1 , avr1)
print("Number greater than average (List 1) : ", mx_avg_1)
mx_avg_2 = greater_avg(list_2 , avr2)
print("Number greater than average (List 2) : ", mx_avg_2)
mx_avg_3 = greater_avg(list_3 , avr3)
print("Number greater than average (List 1) : ", mx_avg_3)
  
```
